# Graph Triangulation: Time Complexity Benchmarking

---

## Overview

This notebook runs comprehensive time complexity benchmarking for the **Biconnected Graph Triangulation Algorithm**. The benchmark suite tests the algorithm's performance against various graph configurations and sizes.

All code, test cases, and results are sourced directly from the GitHub repository: **[TAR2003/Thesis](https://github.com/TAR2003/Thesis)**

---

## 1. Environment Setup

---

Initialize the Python environment and import required libraries for benchmarking and analysis.

```python
import os
import sys
import subprocess
import json
from pathlib import Path
from datetime import datetime

print("✓ Environment setup complete")
print(f"  → Working directory: {os.getcwd()}")
print(f"  → Python version: {sys.version.split()[0]}")
```

## 2. Repository Setup

---

Clone the Thesis repository from GitHub if it doesn't already exist, then navigate to the time_complexity folder.

```python
working_dir = "/kaggle/working"
thesis_dir = os.path.join(working_dir, "Thesis")
time_complexity_dir = os.path.join(thesis_dir, "time_complexity")

print("📋 Repository Setup")
print("=" * 60)

# Check if Thesis folder exists
if os.path.exists(thesis_dir):
    print(f"✓ Thesis folder found at: {thesis_dir}")
else:
    print(f"⏳ Thesis folder not found. Cloning from GitHub...")
    os.chdir(working_dir)
    result = subprocess.run(
        ["git", "clone", "https://github.com/TAR2003/Thesis.git"],
        capture_output=True,
        text=True
    )
    if result.returncode == 0:
        print(f"✓ Successfully cloned Thesis repository")
    else:
        print(f"✗ Error cloning repository: {result.stderr}")
        raise Exception("Failed to clone repository")

# Verify time_complexity folder exists
if os.path.exists(time_complexity_dir):
    print(f"✓ time_complexity folder found at: {time_complexity_dir}")
else:
    print(f"✗ time_complexity folder not found!")
    raise Exception("time_complexity folder not found in Thesis repository")

# Change to time_complexity directory
os.chdir(time_complexity_dir)
print(f"✓ Changed working directory to: {os.getcwd()}")

# List all contents
print(f"\n📁 Contents of time_complexity folder:")
for item in sorted(os.listdir(".")):
    if os.path.isdir(item):
        print(f"  📂 {item}/")
    else:
        print(f"  📄 {item}")

print("=" * 60)
```

## 3. Compilation

---

Compile the C++ benchmarking code with optimizations for accurate performance measurements.

```python
print("\n🔨 Compilation")
print("=" * 60)

# Check if main.cpp exists
if not os.path.exists("main.cpp"):
    print("✗ main.cpp not found!")
    raise Exception("main.cpp not found in time_complexity directory")

print("⏳ Compiling main.cpp with O3 optimization...")
result = subprocess.run(
    ["g++", "-O3", "main.cpp", "-o", "triangulation_benchmark"],
    capture_output=True,
    text=True
)

if result.returncode == 0:
    print("✓ Compilation successful!")
    print(f"  → Output binary: triangulation_benchmark")
    # Check if binary exists and get its size
    if os.path.exists("triangulation_benchmark"):
        size = os.path.getsize("triangulation_benchmark")
        print(f"  → Binary size: {size / 1024:.2f} KB")
else:
    print(f"✗ Compilation failed!")
    print(f"Error: {result.stderr}")
    raise Exception("Compilation failed")

print("=" * 60)
```

## 4. Run Benchmarks

---

Execute the benchmarking binary with real-time output to measure time complexity across various test cases.

```python
print("\n🚀 Benchmarking")
print("=" * 60)
print(f"⏳ Starting benchmark execution at {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
print("=" * 60)
print()

# Run the benchmark with real-time output
result = subprocess.run(
    ["./triangulation_benchmark"],
    cwd=os.getcwd(),
    capture_output=False,
    text=True
)

print()
print("=" * 60)
if result.returncode == 0:
    print(f"✓ Benchmark execution completed successfully")
    print(f"  → Completion time: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
else:
    print(f"✗ Benchmark execution encountered an error (exit code: {result.returncode})")
print("=" * 60)
```

## 5. Results Summary

---

Review the benchmarking results and output files.

```python
print("\n📊 Results Summary")
print("=" * 60)

# Verify results.csv was created during benchmark
if os.path.exists("results.csv"):
    size = os.path.getsize("results.csv")
    print(f"✓ results.csv created ({size} bytes)")
    
    # Read and display the CSV results
    print("\n📈 Benchmark Results (results.csv):")
    print("=" * 60)
    with open("results.csv", "r") as f:
        content = f.read()
        print(content)
    print("=" * 60)
else:
    print("✗ results.csv not found - benchmark may not have completed")

# Check for other output files
output_files = {
    "results.txt": "Text results",
    "results.html": "HTML report"
}

print("\n📁 Additional Output Files:")
for file, description in output_files.items():
    if os.path.exists(file):
        size = os.path.getsize(file)
        print(f"  ✓ {file}: {description} ({size} bytes)")
    else:
        print(f"  ✗ {file}: not found")

print("\n" + "="*60)
print("✓ Benchmarking completed!")
print("="*60)
```
