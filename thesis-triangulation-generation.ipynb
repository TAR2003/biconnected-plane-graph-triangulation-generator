## 🛰️ Step 3: Securely Cloning Private Repository

To bridge the gap between your private GitHub research and the Kaggle environment, we use a **GitHub Fine-Grained Personal Access Token (PAT)**. This allows the Kaggle VM to authenticate securely without exposing your credentials or requiring interactive input.

### 🛡️ Security & Performance Highlights
* **Kaggle Secrets:** We leverage `UserSecretsClient` to pull the token from encrypted storage, keeping your "key" invisible even if the notebook is shared.
* **Authenticated Injection:** The token is injected directly into the HTTPS URL (`https://<TOKEN>@github.com/...`) for seamless non-interactive cloning.
* **Shallow Clone (`--depth=1`):** Optimizes disk space and reduces download time by pulling only the latest snapshot, ignoring the bulky commit history.
* **Fail-Safe Logic:** Includes a 5-minute `subprocess` timeout and comprehensive error handling to prevent the notebook from hanging.

---

### 🛠️ Execution Requirements
| Requirement | Status | Action if Missing |
| :--- | :--- | :--- |
| **Kaggle Internet** | **Required** | Toggle **ON** in the right-hand Settings sidebar. |
| **Kaggle Secret** | **Required** | Add via **Add-ons -> Secrets** with label `thesis-github-repo-secret`. |
| **PAT Scope** | **Required** | Ensure token has **"Contents: Read"** access to the specific repo. |

---



```python
from kaggle_secrets import UserSecretsClient
import subprocess
import os

# 1. Get the token from Kaggle Secrets
user_secrets = UserSecretsClient()
try:
    # Make sure this matches the Label you set in Kaggle Add-ons -> Secrets
    secret_value_0 = user_secrets.get_secret("thesis-github-repo-secret")
except Exception:
    print("✗ Could not find Kaggle Secret. Please check your Secret Label.")
    raise

# Define your repository details
github_user = "TAR2003"
repo_name = "Thesis"
# Inject the token into the URL for authentication
authenticated_url = f"https://{secret_value_0}@github.com/{github_user}/{repo_name}.git"

if not thesis_exists:
    print("\n🌐 Cloning Private Thesis Repository from GitHub")
    print("=" * 70)
    print(f"Repository: https://github.com/{github_user}/{repo_name}.git")
    print(f"Destination: {working_dir}")
    print("=" * 70)
    
    try:
        # Change to working directory first
        os.chdir(working_dir)
        print(f"✓ Changed to working directory: {os.getcwd()}\n")
        
        print("⏳ Cloning in progress (Private Access enabled)...\n")
        
        # Clone with verbose output to show progress
        # Note: We use the authenticated_url here
        result = subprocess.run(
            ["git", "clone", "--progress", "--depth=1", authenticated_url],
            capture_output=False,
            text=True,
            timeout=300  # 5 minute timeout
        )
        
        print("\n" + "=" * 70)
        
        if result.returncode == 0:
            print("✓ Repository cloned successfully!")
            thesis_exists = True
            
            # Show cloned directory size
            if os.path.exists(thesis_dir):
                size = subprocess.run(
                    ["du", "-sh", thesis_dir],
                    capture_output=True,
                    text=True
                )
                print(f"  Size: {size.stdout.split()[0] if size.stdout else 'Unknown'}")
        else:
            print("✗ Clone operation failed!")
            print("Tip: Check if your Fine-Grained token has 'Contents: Read' permission.")
            raise Exception("Git clone failed")
            
    except subprocess.TimeoutExpired:
        print("✗ Clone operation timed out (exceeded 5 minutes)")
        raise Exception("Clone operation timed out")
    except Exception as e:
        print(f"✗ Error during cloning: {str(e)}")
        print("\n⚠️  TROUBLESHOOTING:")
        print("  • Ensure 'Internet' is toggled ON in the Kaggle sidebar")
        print("  • Verify the Token is correct and hasn't expired")
        print("  • Ensure the Secret Label in Kaggle matches 'thesis-github-repo-secret'")
        raise
        
    print("=" * 70)
else:
    print("\n✓ Skipping clone - repository already exists locally")
```

## 3. Build & Compilation

---

### Step 1: Verify Build Requirements

Check that all necessary source files exist before compilation.

```python
print("\n📝 Checking Build Requirements")
print("=" * 70)

required_files = ["main.cpp"]
missing_files = []

for file in required_files:
    if os.path.exists(file):
        size = os.path.getsize(file)
        print(f"  ✓ {file} found ({size} bytes)")
    else:
        print(f"  ✗ {file} NOT found")
        missing_files.append(file)

if missing_files:
    print(f"\n✗ Missing {len(missing_files)} required file(s)!")
    raise Exception(f"Missing files: {', '.join(missing_files)}")
else:
    print(f"\n✓ All required files are present")

print("=" * 70)
```

### Step 2: Compile C++ Code with Optimizations

Compile the main.cpp source file using g++ with O3 optimization flag for maximum performance during benchmarking.

```python
print("\n🔨 C++ Compilation")
print("=" * 70)

compiler = "g++"
source_file = "main.cpp"
output_binary = "triangulation_benchmark"
optimization = "-O3"

print(f"Compiler: {compiler}")
print(f"Source: {source_file}")
print(f"Optimization: {optimization}")
print(f"Output: {output_binary}")
print("\n⏳ Compilation in progress...")
print("=" * 70)

try:
    start_time = datetime.now()
    
    result = subprocess.run(
        [compiler, optimization, source_file, "-o", output_binary],
        capture_output=True,
        text=True,
        timeout=60
    )
    
    end_time = datetime.now()
    compile_time = (end_time - start_time).total_seconds()
    
    print(f"\n✓ Compilation completed in {compile_time:.2f} seconds")
    
    if result.returncode == 0:
        # Check binary exists and get size
        if os.path.exists(output_binary):
            binary_size = os.path.getsize(output_binary)
            print(f"✓ Binary created successfully")
            print(f"  → Filename: {output_binary}")
            print(f"  → Size: {binary_size / 1024 / 1024:.2f} MB")
            print(f"  → Permissions: ", end="")
            result_stat = subprocess.run(["ls", "-lh", output_binary], 
                                        capture_output=True, text=True)
            print(result_stat.stdout.split()[-1])
        else:
            raise Exception("Binary file was not created")
    else:
        print(f"✗ Compilation failed!")
        print(f"\nCompiler output:\n{result.stderr}")
        raise Exception("Compilation failed with errors")
        
except subprocess.TimeoutExpired:
    print("✗ Compilation timed out (exceeded 60 seconds)")
    raise Exception("Compilation timeout")
except Exception as e:
    print(f"✗ Error: {str(e)}")
    raise

print("=" * 70)
```

## 4. Benchmarking Execution

---

### Step 1: Pre-execution Verification

Verify that the compiled binary exists and is executable before starting the benchmark.

```python
print("\n✅ Pre-Execution Verification")
print("=" * 70)

binary_name = "triangulation_benchmark"

if os.path.exists(binary_name):
    # Check if it's executable
    if os.access(binary_name, os.X_OK):
        size = os.path.getsize(binary_name)
        print(f"✓ Binary is ready for execution")
        print(f"  → Filename: {binary_name}")
        print(f"  → Size: {size / 1024 / 1024:.2f} MB")
        print(f"  → Executable: Yes")
    else:
        print(f"⚠ Binary exists but is not executable, fixing...")
        os.chmod(binary_name, os.stat(binary_name).st_mode | 0o111)
        print(f"✓ Binary is now executable")
else:
    print(f"✗ Binary not found: {binary_name}")
    raise Exception("Compiled binary not found. Please run compilation step first.")

print("\n✓ Ready to execute benchmarks")
print("=" * 70)
```

### Step 2: Run Benchmark with Real-Time Output

Execute the triangulation benchmark. The program will process all test cases and write results to results.csv as each test completes. You'll see real-time progress output below.

```python
print("\n🚀 Benchmark Execution")
print("=" * 70)
print(f"Start time: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
print("=" * 70)
print()
print("📊 Running benchmarks with real-time output:\n")

start_time = datetime.now()

try:
    result = subprocess.run(
        ["./triangulation_benchmark"],
        cwd=os.getcwd(),
        capture_output=False,
        text=True
    )
    
    end_time = datetime.now()
    total_time = (end_time - start_time).total_seconds()
    
    print("\n" + "=" * 70)
    print(f"End time: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
    print(f"Total execution time: {total_time // 60:.0f}m {total_time % 60:.1f}s")
    
    if result.returncode == 0:
        print("✓ Benchmark execution completed successfully!")
    else:
        print(f"⚠ Benchmark finished with exit code: {result.returncode}")
    
except Exception as e:
    print(f"\n✗ Error during benchmark execution: {str(e)}")
    raise

print("=" * 70)
```

## 5. Results Analysis

---

### Step 1: Verify Results Files

Check which output files were generated by the benchmark program.

```python
print("\n📁 Checking Generated Output Files")
print("=" * 70)

output_files = {
    "results.csv": "CSV - Results table",
    "results.txt": "Text - Readable results",
    "results.html": "HTML - Interactive report"
}

found_files = {}
total_size = 0

for filename, description in output_files.items():
    if os.path.exists(filename):
        file_size = os.path.getsize(filename)
        found_files[filename] = file_size
        total_size += file_size
        print(f"✓ {filename}")
        print(f"  → {description}")
        print(f"  → Size: {file_size / 1024:.1f} KB")
    else:
        print(f"✗ {filename} (not found)")

print("\n" + "-" * 70)
if found_files:
    print(f"✓ Successfully generated {len(found_files)} file(s)")
    print(f"  Total size: {total_size / 1024 / 1024:.2f} MB")
else:
    print(f"⚠ No output files found - benchmark may not have completed")

print("=" * 70)
```

### Step 2: Display Benchmark Results from CSV

The results.csv file contains all benchmark data written incrementally during test execution. Each row represents one test case with timing measurements.

```python
print("\n📊 Benchmark Results Summary")
print("=" * 70)

csv_file = "results.csv"

if os.path.exists(csv_file):
    try:
        with open(csv_file, "r") as f:
            csv_content = f.read()
        
        lines = csv_content.strip().split('\n')
        print(f"✓ results.csv loaded ({len(lines)} lines)\n")
        
        # Display header and first few rows
        print("📈 CSV Contents:")
        print("-" * 70)
        
        for i, line in enumerate(lines):
            if len(line) > 150:
                # Truncate very long lines for readability
                print(line[:147] + "...")
            else:
                print(line)
            
            # Add separator after header
            if i == 0:
                print("-" * 70)
            
            # Limit display to first 30 rows
            if i >= 29:
                remaining = len(lines) - i - 1
                if remaining > 0:
                    print(f"... and {remaining} more rows")
                break
        
        print("-" * 70)
        
    except Exception as e:
        print(f"✗ Error reading results.csv: {str(e)}")
else:
    print(f"✗ results.csv not found")
    print(f"  → Benchmark may not have completed successfully")
    print(f"  → Check the benchmark execution output above for errors")

print("=" * 70)
```
